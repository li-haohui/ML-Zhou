## 第二章 模型评估与选择

### 经验误差与过拟合

* 错误率(error rate)、精度(accuracy)
* 训练误差(training error)、经验误差(empirical error) ： 发生在训练集上
* 泛化误差(generalization error) ： 发生在新样本上
* 过拟合(overfitting)、欠拟合(undrerfitting)

### 评估方法

* 留出法(hold-out)  
将样本集划分为训练集和测试集

* 交叉验证法(cross validation)  
将数据集 $D$ 划分为 $k$ 个大小相似的互斥子集，然后 $k-1$ 个子集的并集为训练集，余下的子集为测试集。  
特殊情况**留一法**(Leave-One-Out, LOO)：取 $k=m-1$

* 自助法(bootstrapping)  
对有 $m$ 个样本的数据集 $D$，每次有放回地从中抽取一个样本啊，重复执行 $m$ 次。则可得对于一个样本，其在 $m$ 次采样中不被采到的概率为
$$
\lim_{m \to \infty} {\left(1 - \frac{1}{m}\right)^m} = \frac{1}{e} \approx 0.368
$$
由大数定律可得：  
约有 36.8% 的样本未出现在采样数据集 $D'$ 中，于是可以将 $D'$ 作为训练集，将 $D$ \ $D'$ 作为测试集。此时实际评估和期望评估的模型都使用 $m$ 个训练样本。亦称 “包外估计” (out-of-bagestimate)

### 性能度量

* 均方误差(mean squared error)
$$
E(f;D) = \frac{1}{m} \Sigma^{m}_{i=1}(f(x_i) - y_i)^2
$$
更一般形式:
$$
E(f;D) = \int_{x \sim D} (f(x)-y)^2p(x)dx
$$
其中$p(\cdot)$ 为概率密度函数


* 错误率和精度  
 **错误率**：分类错误的样本数占样本总数的比例
$$
E(f;D) = \frac{1}{m} \Sigma^{m}_{i=1} I(f(x_i) \neq y_i)
$$
更一般形式为：
$$
E(f;D) = \int_{x \sim D} I(f(x) \neq y)p(x)dx
$$
其中$p(\cdot)$ 为概率密度函数  
 **精度**: 分类正确的样本数占总样本数的比例
$$
acc(f;D) = \frac{1}{m} \Sigma^{m}_{i=1} I(f(x_i) = y_i)
$$

* 查准率、查全率和F1  
 **查准率**: 查出来的正例中真正例的比例  
 **查全率**: 查出来的正例占实际正例的比例
